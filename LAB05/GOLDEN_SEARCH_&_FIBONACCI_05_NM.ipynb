Golden Rule Search Method

```python
def f_x(x):
  return ((2*(math.sin(x))) - ((x**2)/10))
  # return x**2

def golden_search_method(x_l1,x_u1,e):
  
  x_l,x_u = x_l1,x_u1 
  x_p = x_u1 - 0.618*(x_u1 - x_l1) 
  x_q = x_l1 + 0.618*(x_u1 - x_l1)
  
  E_p = f_x(x_p)
  E_q = f_x(x_q)
  while x_u - x_l > e:  
    if E_p <= E_q:
      x_l = x_l
      x_u = x_q
      x_q = x_p
      x_p = x_u - 0.618*(x_u - x_l)
      E_p = f_x(x_p)
      E_q = f_x(x_q)
    else:
      x_l = x_p
      x_u = x_u
      x_p = x_q
      x_q = x_l + 0.618*(x_u - x_l)
      E_p = f_x(x_p)
      E_q = f_x(x_q)

    # print(f'''
    # x_l : {x_l}
    # x_u : {x_u}
    # x_p : {x_p}
    # x_q : {x_q}
    # E_p : {E_p}
    # E_q : {E_q}''')

  return x_l,x_u
```

```python
x_l1 = 0
x_u1 = 4
e = 0.05
x_l,x_u = golden_search_method(x_l1,x_u1,e)
print(f'''x_l    : {x_l}
f(x_l) : {f_x(x_l)}

x_u    : {x_u}
f(x_u) : {f_x(x_u)}''')
```

Fibonacci Search Method

```python
def fibo(n):
  if n == 0 or n == 1:
    return 1
  else:
    return fibo(n-1)+fibo(n-2)
```

```python
def fibonacci_search_method(x_l1,x_u1,e,n):
  k = 1
  x_l,x_u = x_l1,x_u1 
  x_p = x_u1 - (fibo(n-k)/fibo(n-k+1))*(x_u1 - x_l1) 
  x_q = x_l1 + (fibo(n-k)/fibo(n-k+1))*(x_u1 - x_l1)
  
  E_p = f_x(x_p)
  E_q = f_x(x_q)
  k += 1

  while k <= n-2:
    if E_p <= E_q:
      x_l = x_l
      x_u = x_q
      x_q = x_p
      x_p = x_u - (fibo(n-k)/fibo(n-k+1))*(x_u - x_l)
      E_p = f_x(x_p)
      E_q = f_x(x_q)
    else:
      x_l = x_p
      x_u = x_u
      x_p = x_q
      x_q = x_l + (fibo(n-k)/fibo(n-k+1))*(x_u - x_l)
      E_p = f_x(x_p)
      E_q = f_x(x_q)
    k += 1
    
  # Last two iteration where we need to add 'e' in the n-2th iteration
  for i in range(2):
    if E_p >= E_q:
      x_l = x_l
      x_u = x_q
      x_q = x_p
      x_p = x_u - (fibo(n-k)/fibo(n-k+1))*(x_u - x_l) - e
      E_p = f_x(x_p)
      E_q = f_x(x_q)
    else:
      x_l = x_p
      x_u = x_u
      x_p = x_q
      x_q = x_l + (fibo(n-k)/fibo(n-k+1))*(x_u - x_l) + e
      E_p = f_x(x_p)
      E_q = f_x(x_q)

  return x_l,x_u
```

```python
x_l1 = 0
x_u1 = 4
e = 0.05
n = 20
fx_l,fx_u = fibonacci_search_method(x_l1,x_u1,e,n)
print(f'''x_l    : {fx_l}
f(x_l) : {f_x(fx_l)}

x_u    : {fx_u}
f(x_u) : {f_x(fx_u)}''')
```
